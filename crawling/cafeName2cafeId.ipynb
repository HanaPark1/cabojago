```python
import shutil

# 현재 디렉토리에 있는 .txt 파일 목록 얻기
txt_files = [file for file in os.listdir() if file.endswith(".csv")]

# 파일 이름 변경
for old_name in txt_files:
    new_name = "new_" + old_name
    shutil.move(old_name, new_name)
    print(f"파일 이름 변경: {old_name} -> {new_name}")
```
