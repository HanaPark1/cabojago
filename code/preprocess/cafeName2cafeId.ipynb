### 성수동

```python
import pandas as pd
import numpy as np
import os

# 드라이브 폴더 경로
path = '../data/blog_review/blog_review_result_list/성수동'

# 폴더 내 파일 리스트 가져오기
files = os.listdir(path)
len(files)
# 파일 리스트 출력
# for file in files:
#     print(f"File Name: {file}")

cafeId= pd.read_csv('../data/output_cafelist/cafe_info_list_seongsu.csv')
len(cafeId)

import re
for file in files[10:]:
    try:
        result = re.search(r'성수동 (.*?)_', file)
        
        if result:
            cafeName = result.group(1)
            print(cafeName)
            
            # cafeId에서 '카페이름'이 cafeName을 포함하는 행 찾기
            matching_rows = cafeId[cafeId['카페이름'].str.contains(cafeName)]
            
            if not matching_rows.empty:
                id = int(matching_rows.iloc[0]['카페ID'])
                new_name = "ss_" + str(id) + "_blog.csv"
                
                old_path = os.path.join(path, file)
                new_path = os.path.join(path, new_name)
                
                os.rename(old_path, new_path)
                print(f"파일 이름 변경: {file} -> {new_name}")
            else:
                print(f"{cafeName}에 해당하는 카페ID가 없습니다.")
                old_path = os.path.join(path, file)
                os.remove(old_path)
                
        else:
            print(f"정규표현식 매칭 실패: {file}")
    except Exception as e:
        print(f"에러 발생: {e}")
```
